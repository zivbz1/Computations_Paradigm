```python
# Importing relevant packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, pearsonr
```

```python
# Uploading data
data = "C:\\Users\\zhb4\\Box Sync\\Neural Computations\\Registered Report (Nature Human Behavior)\\Online_Pilot_Results.xlsx"
df = pd.read_excel(data)
df.head()
```

```python
# Creating a figure for decision-making consistency/inconsistency in relation to anxiety severity

# Create a figure with three subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

# Set a common title for all three subplots
fig.suptitle('Decision Inconsistency vs Anxiety Severity', fontweight='bold')

# Set the desired axis ranges
y_min, y_max = -0.05, 1.05 #decision inconsistency
x_min, x_max = 0, 22 #GAD-7 total score

# Scatter plot 1: Anxiety vs Decision Inconsistency Rewards
sns.regplot(data=df, x='Anxiety', y='Decision_Inconsistency_Rewards', ax=axes[0])
axes[0].set_xlabel('GAD-7 Total')
axes[0].set_ylabel('Decision Inconsistency')
axes[0].set_title('Rewards Domain', color='green')
axes[0].set_xlim(x_min, x_max)
axes[0].set_ylim(y_min, y_max)

# Calculate and display correlation coefficient and p-value
corr, p_val = spearmanr(df['Anxiety'], df['Decision_Inconsistency_Rewards'])
axes[0].text(0.4, 0.9, f"Corr: {corr:.3f}\nP: {p_val:.3f}", transform=axes[0].transAxes, color='green')

# Scatter plot 2: Anxiety vs Decision Inconsistency Punishments
sns.regplot(data=df, x='Anxiety', y='Decision_Inconsistency_Punishments', ax=axes[1])
axes[1].set_xlabel('GAD-7 Total')
axes[1].set_ylabel('Decision Inconsistency')
axes[1].set_title('Punishments Domain', color='red')
axes[1].set_xlim(x_min, x_max)
axes[1].set_ylim(y_min, y_max)

# Calculate and display correlation coefficient and p-value
corr, p_val = spearmanr(df['Anxiety'], df['Decision_Inconsistency_Punishments'])
axes[1].text(0.4, 0.9, f"Corr: {corr:.3f}\nP: {p_val:.3f}", transform=axes[1].transAxes, color='red')

# Scatter plot 3: Anxiety vs Decision Inconsistency Mixed
sns.regplot(data=df, x='Anxiety', y='Decision_Inconsistency_Mixed', ax=axes[2])
axes[2].set_xlabel('GAD-7 Total')
axes[2].set_ylabel('Decision Inconsistency')
axes[2].set_title('Mixed Domain',color='orange')
axes[2].set_xlim(x_min, x_max)
axes[2].set_ylim(y_min, y_max)

# Calculate and display correlation coefficient and p-value
corr, p_val = spearmanr(df['Anxiety'], df['Decision_Inconsistency_Mixed'])
axes[2].text(0.4, 0.9, f"Corr: {corr:.3f}\nP: {p_val:.3f}", transform=axes[2].transAxes, color='orange')

# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```
